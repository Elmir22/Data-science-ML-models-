```python
df = df.drop(columns = 'Row ID')
# df.drop(columns = 'Row ID', inplace = True)
```

```python
df
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format = "%d/%m/%Y")
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format = "%d/%m/%Y")
```

```python
df['Order days'] = df['Ship Date'] - df['Order Date']
df
```

```python
df['Order days'].unique()
```

```python
df
```

```python
df['month'] = df['Order Date'].dt.month
df
```

```python
df['year'] = df['Order Date'].dt.year
df
```

```python
df1 = df.groupby(['year','month','Segment']).agg({'Customer ID':'nunique','Order ID':'nunique','Sales':'sum'})
df1.head(10)
```

```python
df1 = df1.reset_index()
df1
```
