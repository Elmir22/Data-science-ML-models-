```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

#Modelling with Pytorch

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import TensorDataset, DataLoader
```

```python
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train.values, dtype=torch.float32).view(-1, 1)
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test.values, dtype=torch.float32).view(-1, 1)
```

```python
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
test_dataset = TensorDataset(X_test_tensor, y_test_tensor)
```

```python
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, shuffle=True)
```

```python
class NeuralNet(nn.Module):
    def __init__(self):
        super(NeuralNet, self).__init__()
        self.fc1 = nn.Linear(X_train_tensor.shape[1], 128)
        self.fc2 = nn.Linear(128, 64)
        self.fc3 = nn.Linear(64,1)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = F.sigmoid(self.fc3(x))
        return x
```

```python
model = NeuralNet()
```

```python
n_epochs = 50
learning_rate = 1e-2
optimizer = optim.SGD(params = model.parameters(), lr=learning_rate)
criterion = nn.BCELoss()
```

```python
def accuracy(y_pred, y_test):
  y_pred_tag = torch.round(y_pred)
  correct_results_sum = (y_pred_tag == y_test).sum().float()
  acc = correct_results_sum/y_test.shape[0]
  acc = torch.round(acc * 100)
  return acc
```

```python
model.train()

for epoch in range(n_epochs):
  epoch_loss = 0
  epoch_acc = 0
  for X_batch, y_batch in train_loader:
    optimizer.zero_grad()
    y_predicted = model(X_batch)

    acc = accuracy(y_predicted, y_batch)
    loss = criterion(y_predicted, y_batch)

    loss.backward()
    optimizer.step()

    epoch_loss += loss.item()
    epoch_acc += acc.item()

  print(f'Epoch {epoch+1}/{n_epochs} | Loss: {epoch_loss/len(train_loader):.5f} | Acc: {epoch_acc/len(train_loader):.3f}')
```

```python
model.eval()

with torch.no_grad():
  torch.no_grad()
  y_predicted = model(X_test_tensor)
  prediction_list = (y_predicted > 0.5).int()
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, f1_score, precision_score, recall_score
```

```python
print(f"Accuracy:{accuracy_score(y_test, prediction_list)}\n --------------------------")
print(f"F1 Score: {f1_score(y_test, prediction_list)}\n --------------------------")
print(f"Precision: {precision_score(y_test, prediction_list)}\n --------------------------")
print(f"Recall: {recall_score(y_test, prediction_list)}\n --------------------------")
print(f"classification report \n{classification_report(y_test, prediction_list)}\n --------------------------")
print(f"confusion_matrix \n{confusion_matrix(y_test, prediction_list)}\n --------------------------")
```
